# Data Cleaning CNTSD and RSUI

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy
import datetime
```

#### CNTSD Dataset

```python
#read CNTSD from file
data = pd.read_excel("2022 Edition CNTSDATA.xlsx", sheet_name="2021 Data")

df_CNTSD = pd.DataFrame(data, )
df_CNTSD = df_CNTSD.drop(0, axis=0)
df_CNTSD
```

```python
#Summary of CNTSD:
print(df_CNTSD.describe())
df_CNTSD.columns
```

#### RSUI Dataset

```python
#read in RSUI data
data = pd.read_excel("RSUI-data-1119.xlsx", sheet_name="Country")

df_rsui = pd.DataFrame(data, )

df_rsui = df_rsui.drop([0,1,2,3,4], axis=0)
df_rsui = df_rsui.drop("Unnamed: 0", axis= 1)
df_rsui.columns = column_names = ["Date","Year", "AFG", "ARE","ARM","AZE","BHR","DJI","DZA","EGY",	"GEO",	"IRN",	"IRQ",	"JOR",	"KAZ",	"KGZ",	"KWT",	"LBN",	"LBY",	"MAR",	"MRT",	"OMN",	"PAK",	"QAT",	"SAU",	"SDN",	"SOM",	"SYR",	"TJK",	"TKM",	"TUN",	"UZB",	"WBG",	"YEM"]
df_rsui
```

##### Summary stats

```python
print(df_rsui.describe())
df_rsui.columns
```

## Cleaning Data
We select only the years and columns we are interested in from CNTSD. We then create two data frames, one made up of all countries and one made up of only the MENA countries contained in the RSUI dataframe. 

```python
#selecting only the years we are interested in; i.e. post-2000
df_CNTSD = df_CNTSD[df_CNTSD["Year"] >= 2000]
df_CNTSD = df_CNTSD[df_CNTSD["Year"] <= 2019]

#keeping only certain columns from CNTSD 
columns_keep = ['Country Code', 'World Bank Code','Country', 'Year', 'Assassinations', 'General Strikes','Guerrilla Warfare','Government Crises', 'Purges', 'Riots', 'Revolutions', 'Anti-Government Demonstrations', 'Weighted Conflict Index', 'Type of Regime', 'Head of State', 'Effective Executive (Type)', 'Effective Executive (Selection)','Degree of Parliamentary Responsibility','Legislative Effectiveness']

df_CNTSD = df_CNTSD[columns_keep]

#selecting only the countries we are intersted in, i.e the MENA countries in the RSUI data set and saving to a new df_CNTSD_MENA to then combine with the RSUI
countries = df_rsui.columns
countries = countries.drop(["Date", "Year"]) 

df_CNTSD_MENA=df_CNTSD[df_CNTSD['World Bank Code'].isin(countries)]

print(df_CNTSD.describe())
df_CNTSD

print(df_CNTSD_MENA.describe())
df_CNTSD_MENA
```

```python
#removing countries without a world bank code. exmaples of such countries are Cyprus (Greek Sector)
df_CNTSD = df_CNTSD[df_CNTSD['World Bank Code'].notna()]

df_CNTSD.isna().sum()
df_CNTSD_MENA.isna().sum()
```

```python
#changing the RSUI dataframe so that it is a similar format to the CNTSD dataframe. 
melted_df = pd.melt(df_rsui, id_vars=["Year", "Date"], var_name="World Bank Code", value_name="RSUI")
melted_df
```

```python
#resetting indices of the dataframes post-cleaning
df_CNTSD = df_CNTSD.reset_index()
df_CNTSD_MENA = df_CNTSD_MENA.reset_index()
melted_df = melted_df.reset_index()

#save cleaned data to a new csv
melted_df.to_csv("RSUI-data-cleaned.csv")
df_CNTSD.to_csv("CNTSD-data-cleaned.csv")
df_CNTSD_MENA.to_csv("CNTSD-MENA-only-data-cleaned.csv")
```
